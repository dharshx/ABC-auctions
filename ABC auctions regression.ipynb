```python
import statsmodels.api as sm 
```

```python
#Multiple Regression Using Statsmodel
#Regression without control variables
y = data['EBITDA%']  # response
X = data[['FTTerm%','PTTerm%','osha/10k consign','lotdamag/consign']]  # predictor
X.head()
```

```python
X = sm.add_constant(X) 
est = sm.OLS(y, X).fit() #OLS Regression
est.summary()
```

```python
fig = sm.graphics.plot_partregress_grid(est)
fig.tight_layout(pad=0.1)
```

```python
#Regression with control variables
a = data['EBITDA%']  # response
b = data[['FTTerm%','PTTerm%','osha/10k consign','lotdamag/consign','type','legacy','sold']]  # predictor
b.head()
```

```python
b = sm.add_constant(b) 
est1 = sm.OLS(a, b).fit() #OLS Regression with control variables
est1.summary()
```

```python
fig = sm.graphics.plot_partregress_grid(est1)
fig.tight_layout(pad=0.01)
```
